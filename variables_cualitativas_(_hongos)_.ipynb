6. Utilizando el método groupby o crosstab, realice una tabla de frecuencias absolutas, relativas y porcentuales para la variable 'CapShape'.

```python
tabla= pd.concat([df.groupby(['CapShape']).size(), df.groupby(['CapShape']).size() / len(df['CapShape']), df.groupby(['CapShape']).size() / len(df['CapShape']) * 100], axis=1)
tabla.columns = ['Frecuencia absoluta', 'Frecuencia relativa', 'Frecuencia porcentual']
tabla
```

```python
# Lista de las columnas que deseas analizar
columnas_a_analizar = ['CapShape', 'CapSurface', 'CapColor', 'Bruises', 'Odor',
                       'GillAttachment', 'GillSpacing', 'GillSize', 'GillColor', 'StalkShape',
                       'StalkRoot', 'StalkSurfaceAboveRing', 'StalkSurfaceBelowRing',
                       'StalkColorAboveRing', 'StalkColorBelowRing', 'VeilType', 'VeilColor',
                       'RingNumber', 'RingType', 'SporePrintColor', 'Population', 'Habitat']

for columna in columnas_a_analizar:
    crosstab=pd.crosstab(df[columna], df['Class'])
    crosstab.plot.bar()
    plt.title(f'Relación entre {columna} y Clase')
    plt.xlabel(columna)
    plt.ylabel('Cantidad')
    plt.xticks(rotation=0)
    plt.show()


```

```python
columnas_a_analizar = ['CapShape', 'CapSurface', 'CapColor', 'Bruises', 'Odor',
                       'GillAttachment', 'GillSpacing', 'GillSize', 'GillColor', 'StalkShape',
                       'StalkRoot', 'StalkSurfaceAboveRing', 'StalkSurfaceBelowRing',
                       'StalkColorAboveRing', 'StalkColorBelowRing', 'VeilType', 'VeilColor',
                       'RingNumber', 'RingType', 'SporePrintColor', 'Population', 'Habitat']

for columna in columnas_a_analizar:
    crosstab = pd.crosstab(df[columna], df['Class'], normalize='index') * 100

    crosstab.plot.bar()
    plt.title(f'{columna} por Clase')
    plt.xlabel(columna)
    plt.ylabel('Porcentaje')
    plt.xticks(rotation=0)
    plt.show()
```
