(h) Si se pudieran hacer infinitas repeticiones del experimento, a qu valor se aproximaran las frecuencias porcentuales mnima y mxima?

Mientras ms lanzamientos se realicen la frecuencia porcentual mnima y mxima estar cada vez ms cerca de su probabilidad inicial es decir 16,6%.

```python
# G -H
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact_manual
conteos = [0, 0, 0, 0, 0, 0]

def lanzar_dado():
    return np.random.randint(1, 7)

def actualizar_grafico(n):
    global conteos
    for _ in range(n):
        result = lanzar_dado()
        conteos[result - 1] += 1
    frecuencias_relativas = [count / sum(conteos) for count in conteos]
    plt.clf()
    plt.bar(np.arange(1, 7), frecuencias_relativas, tick_label=np.arange(1, 7))
    plt.axhline(y=1/6, color='r', linestyle='--')
    plt.text(0.5, 1/6, '1/6     ', color='r', va='center', ha='right')
    plt.ylim(0, 1)
    plt.title('Frecuencias relativas del lanzamiento de un dado')
    plt.xlabel('Cara del dado')
    plt.ylabel('Frecuencia relativa')
    plt.text(5, 0.9, 'n = {}'.format(sum(conteos)), horizontalalignment='right')
    plt.show()
interact_manual(actualizar_grafico,n=(1,1000));
```

(i) modifique el cdigo anterior para hacer la simulacin del lanzamiento de una moneda
(j) considere el siguiente cdigo:
en base a las simulaciones generadas que se puede decir con respecto a los eventos (cara, sello) y (sello, cara)? tienen la misma probabilidad? son el mismo evento?

A medida que los lanzamientos avanzan las probabilidades de que salga cada evento se asemejan a 1/4 de probabilidad, dejar en claro que cara, sello y sello, cara no pertenecen al mismo evento, ya que para obtener cara /sello la primera moneda debe salir cara y la siguiente sello, mientras que en sello/cara lo que ocurre es al revs. Si se realizan menos lanzamientos las probabilidades
 de un evento y otro aumentan y disminuyen.


```python
# i-j
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact_manual
conteos = [0, 0, 0, 0]
def lanzar_moneda():
    return np.random.randint(0, 2)

def actualizar_grafico(n):
    global conteos

    for _ in range(n):
        resultado1 = lanzar_moneda()
        resultado2 = lanzar_moneda()
        caso = (resultado1, resultado2)
        if caso == (0, 0):
            conteos[0] += 1  # (cara, cara)
        elif caso == (0, 1):
            conteos[1] += 1  # (cara, sello)
        elif caso == (1, 0):
            conteos[2] += 1  # (sello, cara)
        elif caso == (1, 1):
            conteos[3] += 1  # (sello, sello)
    probabilidades = [count / sum(conteos) for count in conteos]
    plt.clf()
    etiquetas = ['(Cara, Cara)', '(Cara, Sello)', '(Sello, Cara)', '(Sello, Sello)']
    plt.bar(etiquetas, probabilidades)
    plt.ylim(0, 1)
    plt.axhline(y=1/4, color='r', linestyle='--')
    plt.text(-0.49, 1/4, '1/4     ', color='r', va='center', ha='right')
    plt.title('Probabilidades del lanzamiento de dos monedas')
    plt.xlabel('Combinacin de caras de las monedas')
    plt.ylabel('Probabilidad')
    plt.text(3.5, 0.9, 'n = {}'.format(sum(conteos)), horizontalalignment='right')
    plt.show()
interact_manual(actualizar_grafico, n=(1, 1000))
```
